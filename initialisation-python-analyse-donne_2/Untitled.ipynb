```python
commandes=pd.read_csv('commande.csv')
produits=pd.read_csv('produits.csv')
```

```python
commandes.head()
```

```python
produits.head()
```

```python
datas=pd.merge(produits,commandes,on='id',how='inner')
datas
```

```python
bilan=datas.groupby('id').agg({'prix':'sum','nombre':'count'})
bilan.head()
```

```python
bilan['nombre']
max(bilan['nombre'])
min(bilan['prix'])
```
